```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def read_and_analyze_data(csv_file: str):
    # 读取 CSV 文件
    df = pd.read_csv(csv_file)
    
    # 提取 'credibility' 列
    credibility = df['credibility']
    
    # 统计 'credibility' 列的类别数量
    credibility_counts = credibility.value_counts()  # 统计每个类别的出现次数
    print("Credibility class distribution:")
    print(credibility_counts)
    
    # 绘制 'credibility' 列的分布图
    plt.figure(figsize=(8, 6))
    sns.countplot(x='credibility', data=df, palette="viridis")
    plt.title('Credibility Distribution')
    plt.xlabel('Credibility')
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # 如果类别较多，旋转 x 轴标签
    plt.show()

# 示例：读取 CSV 文件并分析数据
csv_file = "mfbc_media_desc.csv"  # 请替换为你的 CSV 文件路径
read_and_analyze_data(csv_file)
```
