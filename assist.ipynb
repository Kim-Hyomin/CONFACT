```python
import json
import os
from rank_bm25 import BM25Okapi
import nltk
from tqdm import tqdm
import argparse
```

```python
claims_file = '/data/dataset/intermediate_datasets/sample.json'
with open(claims_file, 'r') as f:
    claims = json.load(f)

results = {}


for claim in tqdm(claims):
    claim_text = claim['claim']
    claim_id = claim['id']

    all_sentences = []
    sentence_ids = []

   
    for filename in os.listdir(args.evidence_folder):
        if filename.startswith(str(claim_id)):
            with open(os.path.join(args.evidence_folder, filename), 'r') as f:
                lines = f.readlines()
            
            
                url = lines[0].strip() # Extract the URL from the first line

            
                sentences = f.readlines() 
                sentences = [sentence.strip() for sentence in sentences]  
                all_sentences.extend(sentences)
                sentence_ids.extend([filename] * len(sentences))

  
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in all_sentences]
    bm25 = BM25Okapi(tokenized_sentences)
   
    claim_tokens = nltk.word_tokenize(claim_text)
    top_k = bm25.get_top_n(claim_tokens, all_sentences, n=args.top_k)


    results[claim_id] = top_k


with open(args.output_file, 'w') as f:
    json.dump(results, f)
```
